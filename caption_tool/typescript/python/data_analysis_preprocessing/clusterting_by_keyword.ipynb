# Test 셋 Keyword 분포 count

```python
import json

num = 1
file_path=f"..\\front\\public\\json\splitJson\\split_json_{num}.json"

def read_file(file_path):
    # split JSON 파일 경로
    # JSON 파일 읽기
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    if data is not None:
        key = list(data.keys())[0]
        return data[key]
    
    return None

def extract_synset(file_path):
    data = read_file(file_path)
    data_synsets = []
    for object in data['objects']:
        for synset in object['synsets']:
            data_synsets.append(synset)
            
    data_synset=list(set(data_synsets))
    return data_synset

file_pathes = [f"..\\front\\public\\json\splitJson\\split_json_{i}.json" for i in range(1,2187)]
all_synsets= [extract_synset(file_path) for file_path in file_pathes]
```

```python
import pandas as pd
df_ori = pd.read_csv("[test]immediate_hypernym_count.csv")
df = df_ori.iloc[:,1:]
top100 = df.sort_values("Value",ascending=False).iloc[:100].reset_index(drop=True)
top200 = df.sort_values("Value",ascending=False).iloc[:200].reset_index(drop=True)
top300 = df.sort_values("Value",ascending=False).iloc[:300].reset_index(drop=True)
top500 = df.sort_values("Value",ascending=False).iloc[:500].reset_index(drop=True)
```

```python
import json
from tqdm import tqdm
import time

def read_file(file_path):
    # split JSON 파일 경로
    # JSON 파일 읽기
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    if data is not None:
        key = list(data.keys())[0]
        return data[key]
    
    return None

num = 1
include_100 = []
include_200 = []
include_300 = []
include_500 = []

for num in tqdm(range(1,52994)):
    train_path = f"C:\\workplace\\Kilab\\etri\\dataset\\train\\output\\output\\split_train_{num}.json"
    data = read_file(train_path)
    synsets = [synset['synsets'] for synset in data['objects']]
    unique_synsets = list(set([item for sublist in synsets for item in sublist]))
    
    for synset in unique_synsets:
        if synset in list(top100['Key']):
            include_100.append(train_path)
            include_200.append(train_path)
            include_300.append(train_path)
            include_500.append(train_path)
            
        elif synset in list(top200['Key']):
            include_200.append(train_path)
            include_300.append(train_path)
            include_500.append(train_path)
        
        elif synset in list(top300['Key']):
            include_300.append(train_path)
            include_500.append(train_path
                               )
        elif synset in list(top500['Key']):
            include_500.append(train_path)
            
print("100:",len(list(set(include_100))))
print("200:",len(list(set(include_200))))
print("300:",len(list(set(include_300))))
print("500:",len(list(set(include_500))))
```
