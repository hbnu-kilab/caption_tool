# 평가셋

```python
import json
import os

def clean_human_annotation(human_annotation: str):
    striped_human_annotation = human_annotation.strip()
    result = striped_human_annotation

    # print(striped_human_annotation[:3])
    if striped_human_annotation[:2] == "1." or striped_human_annotation[:2] == "2." or striped_human_annotation[:2] == "3.":
        result = striped_human_annotation[2:].strip()

    return result


for i in range(1, 1+2186):
    file_path = f'C:\\workplace\\Kilab\\etri\\dataset\\취합\\평가셋\\005_core 데이터 검수\\core_{i}.json'
        
    if not os.path.exists(file_path):
        print("파일 없음", i)
    else:    
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
        
            new_data = {}
            VGID = list(data.keys())[0]
            new_data[VGID] = {
                            
                            "regions":data[VGID]["regions"],
                            "narratives":data[VGID]["narratives"],
                            "keywords":data[VGID]["keywords"],
                            "relation_centric_regions":[{"human_annotation": clean_human_annotation(human_annotation["human_annotation"]), 
                                                        "region_ids": human_annotation["box_ids"]} for human_annotation in data[VGID]["relation_centric_regions"]["human_annotations"]]
                            }
            
            new_file_path = f'C:\\workplace\\Kilab\\etri\\dataset\\취합\\평가셋\\006_core 데이터 배포 버전\\core_{i}.json'
            with open (new_file_path, 'w', encoding='utf-8') as make_file:
                    json.dump(new_data, make_file, ensure_ascii=False, indent='\t')
        except:
            print(f"파일 열때 오류: {i}")
```
